# Was ist RGB?

RGB steht f√ºr Rot Gr√ºn/Gelb und Blau. Unser Auge hat Zapfen, womit es verschiedene Farben wahrnehmen kann. Mehr dazu biespielsweise hier: [Video zum Auge (~5 min)](https://)


Dein Bildschrim macht sich genau das zu nutzen. Er besteht aus ganz vielen Pixeln (= kleine Quadrate). Jedes Pixel hat dabei drei Kan√§le:
- einen roten üî¥
- einen gr√ºnen üü¢
- einen blauen üîµ

Auf dem folgenden Bild sieht man mehrfach vergr√∂√üert wie eine Webseite auf einem Bildschirm dargestellt wird. Je mehr man reinzoomt, desto besser erkennt man, dass es viele Quadrate sind, die direkt aneinander gereiht sind. Jedes diser Quadrate, also Pixel, hat die drei genannten Farben. Wenn man genauer hinschaut, dann erkennt man, das die Helligkeit variiert. Zum Beispiel zur Darstellung des schwarzen Textes ist sie geringer, als f√ºr die Darstellung des helleren Hintergrundes.

![](https://referate.mezdata.de/sj2009/lcd_paul-brosinsky/res-wikipedia/Liquid_Crystal_Display_Macro_Example_zoom_2.jpg)

Je nach dem, wie intensiv, also wie hell, jeder einzele Kanal eingestellt wird, so sehen wir eine andere Farbe. Das kannst Du dir vorstellen wie einen Wasserfarbkasten im Kunstunterricht. Wenn du rot und blau mischst, bekommst du violett. Wenn du etwas mehr rot hinzuf√ºgst, wird es eher heller und rosiger. Wenn du aber mehr blau hinzuf√ºgst, dann wird die Farbe eher ein dunkleres Violett.

![](https://cdn.britannica.com/89/234589-050-0E6E9D4B/color-wheel-additive.jpg)

Die Intensit√§ten (= wie viel Farbe gibst du hinzu) lassen sich in RGB in einem Intervall von 0 bis 255 steuern.
- 0 bedeutet, gar nicht
- 255 bedeutet, maximal viel

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_christmas_tree():
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(6, 8))
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-0.5, 1.75)
    ax.axis('off')  # Turn off axes

    # TODO: Die Tanne selbst. Setze die Farben f√ºr die drei Dreiecke:
    erstesDreieck_red = 0
    erstesDreieck_green = 0
    erstesDreieck_blue = 0

    zweitesDreieck_red = 0
    zweitesDreieck_green = 0
    zweitesDreieck_blue = 0

    drittesDreieck_red = 0
    drittesDreieck_green = 0
    drittesDreieck_blue = 0

    tree_colors = [(erstesDreieck_red/255, erstesDreieck_green/255, erstesDreieck_blue/255),
                    (zweitesDreieck_red/255, zweitesDreieck_green/255, zweitesDreieck_blue/255),
                    (drittesDreieck_red/255, drittesDreieck_green/255, drittesDreieck_blue/255)]
    base_widths = [1.5, 1.2, 0.9]
    heights = [0.5, 0.4, 0.3]

    # Start position for the bottom of the tree
    current_y = 0

    for width, height, color in zip(base_widths, heights, tree_colors):
        # Define the triangle vertices
        vertices = [
            (0, current_y + height),  # Top of the triangle
            (-width / 2, current_y),  # Bottom-left
            (width / 2, current_y)  # Bottom-right
        ]
        # Add the triangle as a polygon
        ax.add_patch(patches.Polygon(vertices, closed=True, color=color))
        # Update the current_y to the top of this layer
        current_y += height

    # TODO: Der Baumstamm. Setze hier die RGB Werte f√ºr den Stamm:
    stamm_red = 0
    stamm_green = 0
    stamm_blue = 0

    trunk_color = (stamm_red/255, stamm_green/255, stamm_blue/255)
    ax.add_patch(patches.Rectangle(
        (-0.1, -0.5), 0.2, 0.5, color=trunk_color
    ))

    # TODO: Die Baumspitze bekommt einen Stern. Setze auch hier die Farbe:
    stern_red = 0
    stern_green = 0
    stern_blue = 0
    star_color = (stern_red/255, stern_green/255, stern_blue/255)
    ax.add_patch(patches.RegularPolygon(
        (0, current_y), numVertices=5, radius=0.1, orientation=0, color=star_color
    ))

    # TODO: Schick wird's erst mit Baumschmuck. Hier kannst du sogar drei verschiedene Farben erstellen:
    kugelFarbe1_red = 0
    kugelFarbe1_green = 0
    kugelFarbe1_blue = 0

    kugelFarbe2_red = 0
    kugelFarbe2_green = 0
    kugelFarbe2_blue = 0

    kugelFarbe3_red = 0
    kugelFarbe3_green = 0
    kugelFarbe3_blue = 0

    decoration_colors = [(kugelFarbe1_red/255, kugelFarbe1_green/255, kugelFarbe1_blue/255),
                          (kugelFarbe2_red/255, kugelFarbe2_green/255, kugelFarbe2_blue/255),
                          (kugelFarbe3_red/255, kugelFarbe3_green/255, kugelFarbe3_blue/255)]
    decorations = [
        (0.3, 0.7), (-0.4, 0.6), (0.5, 0.5),
        (-0.3, 0.3), (0.1, 0.2), (-0.5, 0.1)
    ]
    for (x, y), color in zip(decorations, decoration_colors * 2):
        ax.add_patch(patches.Circle((x, y), 0.05, color=color))

    # Display the plot
    plt.show()

# Draw the Christmas tree
draw_christmas_tree()
```
